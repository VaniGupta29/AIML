```python
import matplotlib.pyplot as plt
```

```python
CountOfHousePerYear = df.groupby(['Year_Built']).count()['years_old']
CountOfHousePerYear
```

```python
plt.plot(CountOfHousePerYear.index,CountOfHousePerYear.values)
plt.xlabel("Year")
plt.ylabel("Count of houses built")
plt.title("Count Of House Per Year")
plt.show()
```

```python
plt.bar(CountOfHousePerYear.index,CountOfHousePerYear.values,color='purple')
plt.xlabel("Year")
plt.ylabel("Count of houses built")
plt.title("Count Of House Per Year")
plt.show()
```

```python
QualityOfNeighbourhood = df.groupby(['Neighborhood_Quality']).count()['Square_Footage']
QualityOfNeighbourhood
```

```python
plt.bar(QualityOfNeighbourhood.index,QualityOfNeighbourhood .values,color='skyblue')
plt.xlabel("Quality")
plt.xticks(QualityOfNeighbourhood.index)
plt.ylabel("Count")
plt.title("Quality Of Neighbourhood")
plt.show()
```

```python
plt.pie(QualityOfNeighbourhood.values,labels=QualityOfNeighbourhood.index,autopct='%1.1f%%')
plt.show()
```
