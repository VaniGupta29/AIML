```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\user\\Desktop\\Vani_Linear_Regression\\svm_rats_elephants.csv")
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x="Height_cm", y="Weight_kg", hue="Animal")
plt.title("Rat vs Elephant")
plt.show()
```

```python
df[['Height_cm','Weight_kg','Animal']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE= LabelEncoder()
df['Animal_Encoded']= LE.fit_transform(df['Animal'])
df
```

```python
df[['Height_cm','Weight_kg', 'Animal_Encoded']]
```

```python
X= df[['Height_cm','Weight_kg']]
y = df['Animal_Encoded']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
X = ss.fit_transform(X)
X
```

```python
X = pd.DataFrame(X,columns=("H","W"))
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel="linear")
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
