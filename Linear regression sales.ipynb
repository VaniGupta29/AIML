```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df[['Advertising_Spend']], df['Sales'], test_size=0.3, random_state=200)
print(X_train,X_test,y_train,y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)*100
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Advertising_Spend'], df['Sales'], color='blue', label='Actual Data Points')
plt.plot(X_test, y_pred, color='red', label='Linear Regression Line')
```
